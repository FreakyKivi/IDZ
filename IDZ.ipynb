````python
class parser:
    flag = True
    k = 0
    ch = ''
    formula =""
    n = 0
    
    def start(self):
        self.formula = input('Enter formula: ')
        self.n = len(self.formula)
        self.read_ch()
        self.result = ""
        self.BF()

        if(self.k <= self.n):
            self.flag = False
                
        if(self.flag == False):
            print("Error! Wrong formula.");
        else:
            print("Ok! Formula is correct.");
        
    def read_ch(self):
        if(self.k < self.n+1):
            if(self.k < self.n):
                self.ch = self.formula[self.k];
        else:
            self.flag = False;
        
        self.k += 1;
               
    def BF(self):
        self.FUNC()
        self.EXT()
       
    def EXT(self):
        if ((self.ch == '+') or (self.ch == '-') or (self.ch == '*') or (self.ch == '/')):
            self.O()
            self.BF()
        
    def FUNC(self):
        self.ALPH()
        self.VARS()        
            
    def VARS(self):
        if (self.ch == '('):
            self.read_ch()
            self.FUNC()
            self.EXT_F()
            if(self.ch == ')'):
                self.read_ch()
            else:
                self.flag = False
   
    def EXT_F(self):
        if (self.ch == ','):
            self.read_ch()
            self.FUNC()
            self.EXT_F()
    
    def ALPH(self):
        if((self.ch >= 'a') and (self.ch <= 'z')):
            self.read_ch()
        else:
            self.flag = False
            
    def O(self):
        if((self.ch == '+') or (self.ch == '-') or (self.ch == '*') or (self.ch == '/')):
            self.read_ch()
        else:
            self.flag = False
                
pars = parser()
pars.start()
````
